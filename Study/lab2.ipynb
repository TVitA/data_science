```python
# Импортируем библиотеки
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Создаем датафрейм
seeds_df = pd.read_csv(
"http://qps.ru/jNZUT")
```

```python
# Исключаем информацию об образцах зерна, сохраняем для дальнейшего использования
varieties = list(seeds_df.pop('grain_variety'))
```

```python
# Извлекаем измерения как массив NumPy
samples = seeds_df.values
```

```python
# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(samples, method='complete')
```

```python
# Строим дендрограмму, указав параметры удобные для отображения
dendrogram(mergings,
           labels=varieties,
           leaf_rotation=90,
           leaf_font_size=14,
           )

plt.show()
```

```python
# Импортируем библиотеки
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Определяем модель
dbscan = DBSCAN()

# Обучаем
dbscan.fit(red_wines_df.values)

# Уменьшаем размерность при помощи метода главных компонент
pca = PCA(n_components=2).fit(red_wines_df.values)
pca_2d = pca.transform(red_wines_df.values)

# Строим в соответствии с тремя классами
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')

plt.legend([c1, c2, c3], ['Кластер 1', 'Кластер 2', 'Шум'])
plt.title('DBSCAN нашел 2 кластера и шум')
plt.show()
```

```python
from sklearn.cluster import KMeans

# create kmeans object
kmeans = KMeans(n_clusters = 9) # fit kmeans object to data
kmeans.fit(red_wines_df) # print location of clusters learned by kmeans object
print(kmeans.cluster_centers_) # save new clusters for chart
print(kmeans.labels_)
y_km = kmeans.fit_predict(red_wines_df)
print(y_km)
```
