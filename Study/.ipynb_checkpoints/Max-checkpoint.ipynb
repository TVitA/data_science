```python
# import KMeans
from sklearn.cluster import KMeans

df_new1 = data[['REF','Review_Date','Rating']]

# create kmeans object
kmeans = KMeans(n_clusters = 9)# fit kmeans object to data
kmeans.fit(df_new1)# print location of clusters learned by kmeans object
print(kmeans.cluster_centers_)# save new clusters for chart
print(kmeans.labels_)
y_km = kmeans.fit_predict(df_new1)
print(y_km)

#plt.scatter(df_new1[y_km ==0,0], df_new1[y_km == 0,1], s=100, c='red')
#plt.scatter(df_new1[y_km ==1,0], df_new1[y_km == 1,1], s=100, c='black')
#plt.scatter(df_new1[y_km ==2,0], df_new1[y_km == 2,1], s=100, c='blue')
#plt.scatter(df_new1[y_km ==3,0], df_new1[y_km == 3,1], s=100, c='yellow')
#plt.scatter(df_new1[y_km ==4,0], df_new1[y_km == 4,1], s=100, c='green')
#plt.scatter(df_new1[y_km ==5,0], df_new1[y_km == 5,1], s=100, c='magenta')
#plt.scatter(df_new1[y_km ==6,0], df_new1[y_km == 6,1], s=100, c='pink')
#plt.scatter(df_new1[y_km ==7,0], df_new1[y_km == 7,1], s=100, c='brown')
#plt.scatter(df_new1[y_km ==8,0], df_new1[y_km == 8,1], s=100, c='cyan')
```

```python
#import matplotlib.pyplot as plt

df = pd.DataFrame(kmeans.cluster_centers_, columns=['REF','Review_Date','Rating'])

labels = range(1, 10)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(df['REF'], df['Rating'], label='True Position')

for label, x, y in zip(labels, df['REF'], df['Rating']):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
#from matplotlib import pyplot as plt

linked = linkage(df, 'single')

plt.title('dendrogram')
plt.xlabel('cluster number')
plt.ylabel('distance value')

labelList = range(1, 10)

#plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering

data_agnes = pd.DataFrame(data, columns=['REF','Review_Date','Rating'])

data_agnes_scaled = normalize(data_agnes)

cluster = AgglomerativeClustering(n_clusters=9, affinity='euclidean', linkage='average')

cluster.fit_predict(data_agnes)
```

```python
print('Labels:')
print(cluster.labels_)
print('Number of clusters found:')
print(cluster.n_clusters)

#print(cluster.cluster_centers_)
```

```python
plt.scatter(data_agnes['REF'], data_agnes['Rating'], c=cluster.labels_, cmap='rainbow')
```

```python
plt.scatter(data_agnes['Review_Date'], data_agnes['Rating'], c=cluster.labels_, cmap='rainbow')
```

```python
#from scipy.cluster.hierarchy import dendrogram, linkage

df_agnes = pd.DataFrame(data_agnes, columns=['REF','Review_Date','Rating'])

Z = linkage(df_agnes, 'ward')

plt.title('Hierarchical Clustering Dendrogram')

# Plot axis labels
plt.xlabel('sample index')
plt.ylabel('distance (Ward)')

labelList = range(1, 10)

# Make the dendrogram
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)

# Show the graph
plt.show()
```
