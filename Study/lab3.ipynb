```python
titanic_df_value_features = titanic_df[["PClass", "Age", "Survived", "SexCode"]]
```

```python
titanic_df_value_features.head(5)
```

```python
titanic_df_encoding = pd.get_dummies(titanic_df_value_features[["PClass"]])
```

```python
titanic_df_encoding.head(5)
```

```python
len(titanic_df_encoding)
```

```python
pclass = []

for i in range(len(titanic_df_encoding)):
    if (titanic_df_encoding["PClass_1st"][i] == 1):
        pclass.append(1)
    elif (titanic_df_encoding["PClass_2nd"][i] == 1):
        pclass.append(2)
    else:
        pclass.append(3)
        
titanic_df = titanic_df.drop(['PClass'], axis = 1)        

titanic_df['PClass'] = pclass

titanic_df
```

```python
dataset = titanic_df[["Age", "Survived", "SexCode", "PClass"]]

dataset.head(5)
```

```python
X = dataset.drop(['PClass'], axis = 1)
y = dataset['PClass']
```

```python
X.value_counts()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=36)

clf = DecisionTreeClassifier()


clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
print("  : {:3f}".format(clf.score(X_train,y_train)))
print("  : {:3f}".format(clf.score(X_test, y_test)))
precision = metrics.precision_score(y_test, y_pred, average='micro')
print('Precision: %.3f' % precision)
recall = metrics.recall_score(y_test, y_pred, average='micro')
print('Recall: %.3f' % recall)
score = metrics.f1_score(y_test, y_pred, average='micro')
print('F-Measure: %.3f' % score)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(range(len(y_test)), y_test, '-b',label='Actual')
ax.plot(range(len(y_pred)), y_pred, 'r', label='Predicted')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)
model = RandomForestClassifier(n_estimators = 100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("  : {:3f}".format(clf.score(X_train,y_train)))
print("  : {:3f}".format(clf.score(X_test, y_test)))
precision = metrics.precision_score(y_test, y_pred, average='micro')
print('Precision: %.3f' % precision)
recall = metrics.recall_score(y_test, y_pred, average='micro')
print('Recall: %.3f' % recall)
score = metrics.f1_score(y_test, y_pred, average='micro')
print('F-Measure: %.3f' % score)
```
